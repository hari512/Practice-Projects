```python
## importing all the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
bcd= sklearn.datasets.load_breast_cancer()
```

```python
## displays numpy array
bcd
```

```python
## loading to a pandas dataframe
df=pd.DataFrame(bcd.data,columns=bcd.feature_names)
df
```

```python
## we have only independent features we have to import target feature
df['label']=bcd.target
df.head()
```

```python
## total rows and columns
df.shape
```

```python
## Getting some information about data
df.info()
```

```python
## checking for missing values
df.isnull().sum()
```

```python
## statistical inforamtion about the data
df.describe()
```

```python
## checking the distribution of target variable
df.label.value_counts()
```

1 --> Benign <br>
0 --> Malignant

```python
df.groupby('label').mean()
```

Separing the features and target

```python
x=df.drop(columns='label',axis=1)
y=df.label
```

```python
x
```

```python
y
```

```python
## splitting data into training data and testing data

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

Model Building

```python
model= LogisticRegression()
```

```python
## Training the model with training data

model.fit(x_train,y_train)
```

Model evaluation

```python
## model evaluation

## accuracy of training data
x_train_pred=model.predict(x_train)
training_accuracy= accuracy_score(y_train,x_train_pred)
```

```python
print(f'Accuarcy of training data is : {training_accuracy}')
```

```python
## accuracy of testing data
x_test_pred=model.predict(x_test)
test_accuracy= accuracy_score(y_test,x_test_pred)
```

```python
print(f'Accuarcy of test data is : {test_accuracy}')
```
